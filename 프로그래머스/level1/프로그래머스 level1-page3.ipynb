# 1. 완주하지 못한 선수

```python
def solution1(participant, completion):
    participant.sort()
    completion.sort()
   
    index=0
    for i in range(len(completion)):
        if participant[i]!=completion[i]:
            index=i
            break
        else:
            index+=1
 
    return participant[index]
```

# 2. 성격 유형 검사하기

```python
def solution2(choices, survey):
    info=[[0 for _ in range(2)]for _ in range(4)]
    ptype=[['R','T'],['C','F'],['J','M'],['A','N']]
    answer = ''
    for i in range(len(survey)):
        cnt=0
        if choices[i][0]=='R' and choices[i][1]=='T':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[0][0]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[0][1]+=cnt
                    
        if choices[i][0]=='T' and choices[i][1]=='R':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[0][1]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[0][0]+=cnt
                
        if choices[i][0]=='C' and choices[i][1]=='F':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[1][0]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[1][1]+=cnt
                    
        if choices[i][0]=='F' and choices[i][1]=='C':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[1][1]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[1][0]+=cnt
                
        if choices[i][0]=='J' and choices[i][1]=='M':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[2][0]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[2][1]+=cnt
                    
        if choices[i][0]=='M' and choices[i][1]=='J':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[2][1]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[2][0]+=cnt
        
                
        if choices[i][0]=='A' and choices[i][1]=='N':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[3][0]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[3][1]+=cnt
                    
        if choices[i][0]=='N' and choices[i][1]=='A':
            if survey[i]==4:
                cnt=0
            elif survey[i]<4:
                cnt+=4-survey[i]
                info[3][1]+=cnt
            elif survey[i]>4:
                cnt+=survey[i]-4
                info[3][0]+=cnt
                
    i=[]
    for z in info:
        if z[0]<z[1]:
            i.append(1)
        else:
            i.append(0)
    for p in range(4):
        answer+=ptype[p][i[p]]
        
    return answer
```

# 3. 체육복

```python
def solution3(n, lost, reserve):
   
    s_reserve=set(reserve)-set(lost)
    s_lost=set(lost)-set(reserve)
   
    for i in s_reserve:
        if i-1 in s_lost:
            s_lost.remove(i-1)
           
        elif i+1 in s_lost:
            s_lost.remove(i+1)
         
    return n-len(s_lost)
```

# 4. 키패드 누르기

```python
def solution4(numbers, hand):
    answer = ''
    #왼손잡이부터
    row=0
    col=0
    left=location('*')
    right=location('#')
    for i in numbers:
        if i==1 or i==4 or i==7:
            answer+='L'
            left=location(i)
        elif i==3 or i==6 or i==9:
            answer+='R'
            right=location(i)
        else:
            loc=location(i)
            
            #거리계산을 어떻게 할까
            left_dis=abs(left[0]-loc[0])+abs(left[1]-loc[1])
            right_dis=abs(right[0]-loc[0])+abs(right[1]-loc[1])
            
            if left_dis<right_dis:
                answer+='L'
                left=loc
            elif left_dis>right_dis:
                answer+='R'
                right=loc
            else:
                if hand=='right':
                    answer+='R'
                    right=loc
                else:
                    answer+='L'
                    left=loc
    return answer

def location(n):
    loc={1:[0,0],2:[0,1],3:[0,2],
        4:[1,0],5:[1,1],6:[1,2],
        7:[2,0],8:[2,1],9:[2,2],
        '*':[3,0],0:[3,1],'#':[3,2]}
    
    return loc[n]
```

# 5. 크레인 인형뽑기 게임

```python
def solution5(board, moves):
    answer=0
    stack=[]
    l=len(board)

    for index in range(len(moves)):
        for i in range(l):
            #moves[index]-1열을 다니며 0이면 다음 행으로 넘어감
            if board[i][moves[index]-1]!=0:
                #추가
                stack.append(board[i][moves[index]-1])
                #0으로 다시 설정
                board[i][moves[index]-1]=0
                #스택의 원소 하나가 들어가므로 개수 추가
                #만약 원소의 개수가 1보다 많을시 비교
                if len(stack)>1:
                    #같은것이 잇으면 둘다 pop
                    if stack[-1]==stack[-2]:
                        stack.pop()
                        stack.pop()
                        answer+=1
                        answer+=1

                break
       
    return answer 
```

# 6. 신고 결과 받기

```python
from collections import defaultdict

def solution5(id_list, report, k):
    #중복 제거
    report=set(report)
    answer=[]
    
    #신고한 사람
    user=defaultdict(set)
    #신고한 횟수
    report_info=defaultdict(int)
        
    for i in report:
        reporter=i.split(" ")[0]
        reported=i.split(" ")[1]
        user[reporter].add(reported)
        report_info[reported]+=1
        
    for i in id_list:
        cnt=0
        for u in user[i]:
            if report_info[u]>=k:
                cnt+=1
        answer.append(cnt)
    
    print(user)
    print(report_info)
    return answer
```
